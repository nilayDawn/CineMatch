Step 1: Import the ast module
import ast

The ast module provides the function literal_eval() which safely converts a string that looks like a Python object (like a list or dictionary) into an actual Python object.

Example:

"[{'id': 28, 'name': 'Action'}]"

becomes

[{'id': 28, 'name': 'Action'}]
Step 2: Apply a function to the genres column
df['genres'].apply(...)

The apply() function runs a given function on each row of the column.

Step 3: Use a lambda function
lambda x:

A lambda function is a small anonymous function.
Here, x represents each value in the genres column.

Example value of x:

"[{'id': 28, 'name': 'Action'}, {'id': 12, 'name': 'Adventure'}]"
Step 4: Convert string to Python list
ast.literal_eval(x)

This converts the string into a Python list of dictionaries.

Example result:

[
 {'id': 28, 'name': 'Action'},
 {'id': 12, 'name': 'Adventure'}
]
Step 5: Extract genre names
[i['name'] for i in ast.literal_eval(x)]

This is list comprehension. It loops through each dictionary and extracts the name value.

Result:

['Action', 'Adventure']
Step 6: Join the names into one string
" ".join(...)

The join() function combines the list into a single string separated by spaces.

Result:

"Action Adventure"

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #remove the punctuations
  text = re.sub('[^a-zA-Z]', ' ', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]
  #lemitization
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df = df.reset_index(drop= True)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[ : :-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```
